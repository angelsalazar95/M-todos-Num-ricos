```python
import math
import matplotlib.pyplot as plt
from tabulate import tabulate
#%matplotlib inline
time,xu,xl,fc1,fc2,vv,EB,ERFP,EPF,ENR,ErS,ErSm=[],1.0,0.0,1,1,0.567143,[],[],[],[],[],[]
def func(num):
    fxb=(math.exp(-num)-num)
    return fxb

for i in range (10):
    xr=(xl+xu)/2
    fc1=func(xl)
    fc2=func(xr)
    if ((fc1*fc2)<0):
        xu=xr
    if((fc1*fc2)>0):
        xl=xr
    Erpvb=math.fabs(((vv-xr)/(vv))*100)
    EB.append(Erpvb)
    time.append(i)

xu1,xl1,fc11,fc22=1.0,0.0,1,1
vv=0.567143
for j in range (10):
    xr1=xu1-((funcion(xu1)*(xl1-xu1))/(funcion(xl1)-(funcion(xu1))))
    fc11=funcion(xl1)
    fc22=funcion(xr1)
    if (fc11*fc22)<0:
        xu1=xr1
    if (fc11*fc22)>0:
        xl1=xr1
    Erpvf=math.fabs(((vv-xr1)/(vv))*100)
    ERFP.append(Erpvf)

fun=0
for k in range (10):
    Erpvp=math.fabs(((vv-fun)/(vv))*100)
    EPF.append(Erpvp)
    fun=math.exp(-fun)


fun1,ENR=-1,[]
for l in range (10):
    fun1=fun1-((math.exp(-fun1)-fun1)/(-math.exp(-fun1)-1))
    Erpvn=math.fabs(((vv-fun1)/(vv))*100)
    ENR.append(Erpvn)


xmenos1,x0,ErS,t1=0.0,1.0,[],[]
for m in range (7):
    fun2=x0-(((math.exp(-x0)-x0)*(xmenos1-x0)))/((math.exp(-xmenos1)-xmenos1)-(math.exp(-x0)-x0))
    xmenos1=x0
    x0=fun2
    Erpvs=math.fabs(((vv-fun2)/(vv))*100)
    ErS.append(Erpvs)
    t1.append(m)
    

xo,s,ErSm=1.0,0.01,[]
k=(xo+(s*xo))
for n in range (10):
    fmas1=xo-((s*xo)*((math.exp(-xo))-xo))/((math.exp(-k)-k)-(math.exp(-xo)-xo))
    xo=fmas1
    k=(xo+(s*xo))
    Erpvsm=math.fabs(((vv-fmas1)/(vv))*100)
    ErSm.append(Erpvsm)
print "Como podemos observar de los diferentes métodos implementados el cual desciendente con \
mayor rápidez en su error relativo\nporcentual verdadero es el de la Secante Modificada. "
plt.plot (time, EB, linewidth=2.0, linestyle='--', label='Biseccion')
plt.plot(time,ERFP, linewidth=3.0, linestyle='--',label='Falsa Posicion')
plt.plot(time,EPF, marker='o',linestyle='--',label='Punto Fijo')
plt.plot(time,ENR, marker='*', linestyle='-.',label='Newton-Raphson')
plt.plot(t1,ErS, linewidth=2.0, label='Secante')
plt.plot(time,ErSm, linewidth=2.0, label='Secante Modificada')
plt.xlabel('Iteracion')
plt.ylabel('ERPV')
plt.grid(True)
plt.legend()
plt.show()
```
