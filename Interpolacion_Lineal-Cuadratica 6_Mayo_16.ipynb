```python
#lineal
import matplotlib.pyplot as plt
import math
%matplotlib inline
t,log, vec=[],[],[]
x,x0, x1,fx0,fx1=2.0,1.0,6.0,0.0, math.log(6)
x1b,fx1b=4.0, math.log(4)
x2,fx2,b0=6.0, math.log(6),fx0


f12a= fx0 + ((fx1-fx0)/(x1-x0))*(x-x0)
f12b = fx0 + ((fx1b-fx0)/(x1b-x0))*(x-x0)
b1=((fx1b-fx0)/(x1b-x0))
b2=((((fx2-fx1b)/(x2-x1b))-(b1))/(x2-x0))

f22=(b0+(b1*(x-x0))+b2*(x-x0)*(x-x1b))


arr1=[fx0,f12a,fx1]
arr2=[fx0,f12b,fx1b]
arr3=[fx0,f22,fx1b]

t1=[1,2,6]
t2=[1,2,4]
print"\nEstimaciones con interpolaciones"
print "X1=6 :",f12a
print "X1=4: ",f12b
print "Cuadratica X1=4: ",f22
for i in range (1,8):
    vec.append((b0+(b1*(i-x0))+b2*(i-x0)*(i-x1b)))
    t.append(i)
    log.append(math.log(i))
    
plt.title('Interpolacion lineal y cuadratica')
plt.plot(t,log, marker='+', linewidth=2.0,linestyle='--',label='f(x)= ln(2)')
plt.plot(t1,arr1, marker='o',linewidth=2.0 , linestyle='-.',label='Estimacion x1=6')
plt.plot(t2,arr2,marker='v',linewidth=2.0 , linestyle='-.',label='Estimacion x1=4')
plt.plot(t,vec, marker='s',linewidth=2.0 , linestyle='-.')
plt.plot(2,f22, 'p', label='f2(2)=0.566')
plt.axis([0,8,-1,3])
plt.xlabel('X')
plt.ylabel('f(X)')
plt.legend(loc='upper left', numpoints=1)
plt.grid()
plt.show()
```
