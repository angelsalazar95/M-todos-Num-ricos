```python
import matplotlib.pyplot as plt
from tabulate import tabulate
%matplotlib inline
xi=[1,2,3,4,5,6,7]
yi=[0.5,2.5,2.0,4.0,3.5,6.0,5.5]
tab=[]
n=len(xi)
mult, sum1,sumax,sumay,t,z=0,0,0,0,[],[]

for i in range (n):
    sumax+=xi[i]
    sumay+=yi[i]
    mult+=xi[i]*yi[i]
    sum1+=xi[i]**2

expo=sumax**2
a1=(n*mult-(sumax*sumay))/(n*sum1-expo)
xmedia=sumax/n
ymedia=sumay/n
a0=ymedia-(a1*xmedia)
for j in range (n+1):
    y=a0+a1*j
    t.append(j)
    z.append(y)
    
for k in range(n):
    tab.append([xi[k],yi[k],((yi[k]-ymedia)**2),((yi[k]-a0-(a1*xi[k]))**2)])
    
print tabulate(tab,headers=["xi", "yi","(y-y~)^2","(yi-a0-a1*xi)^2"],tablefmt='pipe')
plt.title("Regresion por Minimos Cuadrados")

plt.plot(t,z, '-', label='recta y=0,07142857+0,8392857x')
plt.plot(xi,yi, '*', label='xi,yi')
plt.ylabel('Ajuste minimos cuadrados')
plt.xlabel('(x)')
plt.axis([0,8,0,9])
plt.legend(loc='upper left', numpoints=1)
plt.grid()
plt.show()
```
